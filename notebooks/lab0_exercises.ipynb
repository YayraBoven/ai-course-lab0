```python
"""
Exercise 5: Pandas Data Analysis
Analyze a dataset of student performance.
"""

import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(42)
n_students = 200

data = {
    'student_id': range(1000, 1000 + n_students),
    'major': np.random.choice(['CS', 'Math', 'Physics', 'Biology'], n_students),
    'year': np.random.choice([1, 2, 3, 4], n_students),
    'exam_score': np.random.normal(75, 10, n_students).clip(0, 100),
    'assignments_completed': np.random.randint(0, 11, n_students),
    'hours_studied': np.random.normal(15, 5, n_students).clip(1, 40)
}

df = pd.DataFrame(data)

# Introduce some NaN values
df.loc[np.random.choice(n_students, 10), 'exam_score'] = np.nan
df.loc[np.random.choice(n_students, 5), 'hours_studied'] = np.nan

# Task 1: Data Cleaning and Exploration (10 points)
# TODO: Display basic information about the dataset

print(df)
print(df.head(100))
print(df.tail(50))
print(df.columns)
print(df.dtypes)

# TODO: Identify and count missing values
print(df.isnull().sum())
#This means exam_score and hours_studied have 10 and 5 missing/empty values, respectively.


```

```python
# TODO: Fill missing exam_score with the mean score for the student's major
df['exam_score'] = df['exam_score'].fillna(df.groupby('major')['exam_score'].transform('mean'))

# TODO: Fill missing hours_studied with the median for the student's year
df['hours_studied'] = df['hours_studied'].fillna(df.groupby('year')['hours_studied'].transform('median'))

print(df.isnull().sum())
#print(df)
```

```python

# Task 2: Analysis (10 points)
# TODO: Calculate and display the average exam_score by major
print(df.groupby('major')['exam_score'].mean())

# TODO: Find the major with the highest average exam_score
print(df.groupby('major')['exam_score'].mean().idxmax())

# TODO: Calculate the correlation between hours_studied and exam_score
print(df['hours_studied'].corr(df['exam_score']))
# A value of -0.055981383472924404 was gotten for the correlatipn between the hours studies and exam score. This values shows a very weak negative relationship between these
#two variables. This means both the hours studied and exam scores are not related in any way.

# TODO: Create a new column 'performance' with categories:
#       'Excellent' (>90), 'Good' (80-90), 'Average' (70-80), 'Needs Improvement' (<70)
#To do this, a function would be created, taking the input of "scores" so that an accurate output description of the grade for a student would be given.
def grade_description(grade):
  if grade > 90:
    return 'Excellent'
  elif grade >= 80:
    return 'Good'
  elif grade >= 70:
    return 'Average'
  else:
    return 'Needs Improvement'
#After creating this function, we would use the panda method/function .apply(). This takes the function we have created and runs it through every value in the exam_score column.
#While using the .apply() on the exam_score values in its column, a performance column would be created as stated in the question. Below is the implementation:
df['performance'] = df['exam_score'].apply(grade_description)
print(df)
```

```python
# Task 3: Advanced Analysis (10 points)
# TODO: For each major and year combination, calculate:
#       - Number of students
#       - Average exam score
#       - Average hours studied

df.groupby(['year','major']).agg(num_stu = ('student_id','count'), avg_score = ('exam_score','mean'), avg_hours = ('hours_studied','mean'))


# TODO: Identify top 5 students based on exam_score (handle ties appropriately)
sorted_values = df.sort_values('exam_score', ascending=False)
print(sorted_values)
sorted_values.head(5)

#The above method used could not handle the ties, therefore we would have to create a new column (position),where we use the .rank() method to assign a position to each value.
df['position'] = df['exam_score'].rank(method='min', ascending=False)
first_students = df[df['position'] <= 5]
print(first_students)

# TODO: Create a pivot table showing average exam_score by major (rows) and year (columns)
mean_scores = df.pivot_table(values= 'exam_score', index= 'major', columns='year', aggfunc= 'mean')
print(mean_scores)


```

```python
"""
Exercise 7: Integration Challenge
Combine NumPy, Pandas, and Matplotlib to solve a mini data science problem.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Scenario: You're analyzing customer data for an e-commerce company.
# Generate synthetic data and provide insights.

np.random.seed(42)
n_customers = 500

# Generate customer data
ages = np.random.randint(18, 70, n_customers)
income = np.random.normal(50000, 20000, n_customers).clip(15000, 150000)
purchase_freq = np.random.poisson(5, n_customers)
avg_purchase_value = np.random.normal(100, 30, n_customers).clip(10, 500)

# Create DataFrame
customers = pd.DataFrame({
    'age': ages,
    'income': income,
    'purchase_frequency': purchase_freq,
    'avg_purchase_value': avg_purchase_value
})


# TODO: Calculate customer lifetime value (CLV)
#CLV = purchase_frequency * avg_purchase_value * (1 + churn_risk)
#where churn_risk = 1 - (purchase_frequency / max_frequency)

max_frequency = customers['purchase_frequency'].max()
customers['churn_risk'] = 1 - (customers['purchase_frequency'] / max_frequency)
customers['CLV'] = customers['purchase_frequency'] * customers['avg_purchase_value'] * (1 + customers['churn_risk'])
customers.head(5)
print(customers)

# TODO: Create age groups: 18-25, 26-35, 36-50, 51-70
customers['age_group'] = pd.cut(
    customers['age'],
    bins= [18,25,35,50,70],
    labels = ['18-25', '26-35', '36-50', '51-70']
)
print(customers['age_group'])
customers.head(5)

# TODO: For each age group, calculate:
#       - Number of customers
#       - Average income
#       - Average CLV
#       - Total CLV

print(customers.groupby(['age_group']).agg(num_customers = ('age','count'), average_income = ('income','mean'), average_CLV = ('CLV','mean'), total_CLV = ('CLV', 'sum') ))


# TODO: Identify top 10% of customers by CLV
#With this question, since we want the top 10% we would find the value of the 90th percentile, and look for customers whose CLV is greater than this value to be found.
border_line = customers['CLV'].quantile(0.90)
print(border_line)

top_cus = customers[customers['CLV'] >= border_line]
print(top_cus)

# TODO: Create visualizations:
#       1. Scatter plot of income vs CLV (color by age group)
#       2. Bar chart of average CLV by age group
#       3. Correlation heatmap


#       1. Scatter plot of income vs CLV (color by age group)
plt.figure(figsize=(10,7))
sns.scatterplot(data=customers, x='income', y='CLV', hue='age_group')
plt.title('The Income and CLV of Customers (Based of Age Groups)')
plt.show()

#       2. Bar chart of average CLV by age group
#Here we would start by grouping customers by their age group and we would move on to find each age_group CLV's average/mean.
plt.figure(figsize=(10,7))
avg_CLV = customers.groupby('age_group')['CLV'].mean().reset_index()
plt.title('Average CLV by Age Group')
sns.barplot(data=avg_CLV, x= 'age_group', y="CLV")



#       3. Correlation heatmap
#Inner/ Second square bracket is for the list of the columns not mere selections of the columns
plt.figure(figsize=(10,7))
sns.heatmap((customers[['age', 'income', 'purchase_frequency', 'avg_purchase_value', 'CLV', 'churn_risk']]).corr(),
            annot=True,
            cmap='plasma'

)
plt.title('Correlation Heatmap for All Attributes')







```

Also the correlation heatmap confirms that income is not the factor that drives their CLV since it is near the number 0 (0.014), but the factor that drives their CLV comes from how frequently they shop which is evident in the heatmap as 0.69 correlation with the CLV and purchase frequency.
